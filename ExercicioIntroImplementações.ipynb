# Introdução às Implementações

## Importações

```python
import pandas as pd
import os
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
```

## Dados de Entrada

## Arquivos

```python
pronto = False
documentos = []
file = '/content/dataset1/documentos'
while pronto == False:
    for filename in os.listdir(file):
        if filename.endswith(".txt"):
            with open(os.path.join(file, filename), 'r') as f:
                documentos.append(f.read())
    if file == '/content/dataset1':
      pronto = True
    file = '/content/dataset1'
print(documentos)
documentos.sort()
print(documentos)
```

Após sort:

0 -> separadores

1 -> doc2

2 -> doc4

3 -> doc1

4 -> doc3

5 -> doc5

6 -> stopwords

7 -> consulta

```python
textos = [documentos[3], documentos[1], documentos[4], documentos[2], documentos[5]]
print(textos)
```

## Separação

```python
separadores = documentos[0].split('\n')
print(separadores)
for i in range(len(textos)):
  for j in range(1,len(separadores)):
    textos[i] = textos[i].replace(separadores[j],separadores[0])
    textos[i] = textos[i].replace(separadores[0]*2,separadores[0])
    if textos[i].endswith(separadores[0]):
      textos[i] = textos[i][:-1]
    if textos[i].startswith(separadores[0]):
      textos[i] = textos[i][1:]
  textos[i] = textos[i].split(separadores[0])
  print(textos[i])
```

## Normalização

```python
for i in range(len(textos)):
  for j in range(len(textos[i])):
    textos[i][j] = textos[i][j].lower()
print(textos)
```

## Eliminação de Stopwords

```python
nltk.download('stopwords')
stopwords_pt = set(nltk.corpus.stopwords.words('portuguese'))
```

```python
for palavra in stopwords_pt:
  for i in range(len(textos)):
    while palavra in textos[i]:
      textos[i].remove(palavra)
print(textos)
```

## Aplicação de Stemming

```python
stemmer = SnowballStemmer("portuguese")
for i in range(len(textos)):
  for j in range(len(textos[i])):
    textos[i][j] = stemmer.stem(textos[i][j])
print(textos)
```

## Indexação

```python
palavras = []
for i in range(len(textos)):
  for j in range(len(textos[i])):
    if textos[i][j] not in palavras:
      palavras.append(textos[i][j])
print(palavras)
```

```python
dados = []
for i in range(len(palavras)):
  frequencias = []
  for j in range(len(textos)):
    frequencias.append(textos[j].count(palavras[i]))
  dados.append(pd.Series(frequencias))

dataFrameFrequencia=pd.DataFrame({'peã': dados[0], 'caval': dados[1], 'pec': dados[2], 'xadrez': dados[3], 'melhor': dados[4], 'jog': dados[5], 'envolv': dados[6], 'torr': dados[7], 'rei': dados[8], 'lac': dados[9], 'boi': dados[10], 'rodei': dados[11], 'polic': dados[12]})
print(dataFrameFrequencia)
```

# Modelo Booleano

## AND entre todos os elementos da consulta

```python
consulta = documentos[7].split()
consulta_indices = []
for i in range(len(consulta)):
  consulta[i] = stemmer.stem(consulta[i])
  consulta_indices.append(palavras.index(consulta[i]))
consulta_indices.sort()
resultado = []
print(consulta)
for i in range(len(textos)):
  arquivo_tem = 1
  for j in range(len(consulta)):
    arquivo_tem = arquivo_tem * dados[consulta_indices[j]][i]
  resultado.append(arquivo_tem)
print(resultado)
```

```python
display_resultado = []
for i in range(len(resultado)):
  if resultado[i] != 0:
    display_resultado.append(i)
if display_resultado == []:
  print("Nenhum documento cumpre todos os requerimentos da pesquisa.")
else:
  print("Os documentos resultantes da consulta são:")
  for i in range(len(display_resultado)):
    print(f"doc{display_resultado[i]+1}.txt")
```

## OR entre todos os elementos da consulta

```python
consulta = documentos[7].split()
consulta_indices = []
for i in range(len(consulta)):
  consulta[i] = stemmer.stem(consulta[i])
  consulta_indices.append(palavras.index(consulta[i]))
consulta_indices.sort()
resultado = []
print(consulta)
for i in range(len(textos)):
  arquivo_tem = 0
  for j in range(len(consulta)):
    arquivo_tem = arquivo_tem + dados[consulta_indices[j]][i]
  resultado.append(arquivo_tem)
print(resultado)
```
