# Ponderação de Termos

## Frequência de Termos

```python
term_freq = []
for i in range(len(dados)):
  frequencias = []
  for j in range(len(dados[0])):
    tftemp = dados[i][j]
    if tftemp > 0:
      tftemp = 1 + math.log2(tftemp)
    frequencias.append(tftemp)
  term_freq.append(frequencias)
print(term_freq)
```

## Frequência Inversa de Documentos

```python
inv_freq = []
for i in range(len(dados)):
  invtemp = 0
  for j in range(len(dados[0])):
    if dados[i][j] > 0:
      invtemp += 1
  if invtemp > 0:
    freq_inv = math.log2(len(dados[0])/invtemp)
  else:
    freq_inv = 0
  inv_freq.append(freq_inv)
print(inv_freq)
```

## Ponderação TF-IDF

```python
tf_idf = []
for i in range(len(dados)):
  tf_idf_temp = []
  for j in range(len(dados[0])):
    tf_idf_temp.append(term_freq[i][j] * inv_freq[i])
  tf_idf.append(tf_idf_temp)
print(tf_idf)
```

# Modelo Vetorial

## Vetores

```python
vetores = []
for i in range(len(dados[0])):
  vetor = []
  for j in range(len(dados)):
    vetor.append(tf_idf[j][i])
  vetores.append(vetor)
print(vetores)
```

## Consulta

```python
consulta_dados = []
consulta_vetor = []
for i in range(len(dados)):
  consulta_dados.append(consulta.count(palavras[0][i]))
  consulta_temp = consulta_dados[i]
  if consulta_temp > 0:
    consulta_temp = (1 + math.log2(consulta_temp)) * inv_freq[i]
  consulta_vetor.append(consulta_temp)
print(consulta_dados)
print(consulta_vetor)
```

## Similaridade com a Consulta e Normalização

```python
similaridade = []
for i in range(len(vetores)):
  modulo_doc = 0
  prod_int = 0
  modulo_consulta = 0
  for j in range(len(consulta_vetor)):
    prod_int += vetores[i][j] * consulta_vetor[j]
    modulo_doc += vetores[i][j]**2
    modulo_consulta += consulta_vetor[j]**2
  modulo_doc = math.sqrt(modulo_doc)
  modulo_consulta = math.sqrt(modulo_consulta)
  similaridade.append(prod_int/(modulo_doc * modulo_consulta))
print(similaridade)
```

## Ranking

```python
ranking = []
rank_temp = similaridade
for i in range(len(dados[0])):
  maior_rank = -10
  maior_index = 0
  for j in range(len(rank_temp)):
    if rank_temp[j] > maior_rank:
      maior_rank = rank_temp[j]
      maior_index = j
  ranking.append(maior_index)
  rank_temp[maior_index] = -10
print(ranking)
print("Lista dos documentos ordenados por ranking:")
for i in range(len(ranking)):
  print(f"doc{ranking[i]+1}.txt")
```

```python
docs_com_termo = []
for i in range(len(dados)):
  temp = 0
  for j in range(len(dados[0])):
    if dados[i][j] > 0:
      temp += 1
  docs_com_termo.append(temp)
print(docs_com_termo)
```

```python
tam_doc = []
for i in range(len(dados[0])):
  tam = 0
  for j in range(len(dados)):
    tam += dados[j][i]
  tam_doc.append(tam)
print(tam_doc)
```

```python
tam_doc_media = 0
for i in range(len(tam_doc)):
  tam_doc_media += tam_doc[i]
tam_doc_media /= len(tam_doc)
print(tam_doc_media)
```

## Similaridade BM25

```python
similaridadeBM25 = []
for i in range(len(dados[0])):
  simBM25 = 0
  for j in range(len(dados)):
    if consulta_dados[j] > 0 and dados[j][i] > 0:
      Bij = (K1 + 1) * dados[j][i]/(K1 * (1 - b + b * tam_doc[i]/tam_doc_media) + dados[j][i])
      simBM25 += Bij * math.log2((len(dados[0]) - docs_com_termo[j] + 0.5)/(docs_com_termo[j] + 0.5))
  similaridadeBM25.append(simBM25)
print(similaridadeBM25)
```

## Ranking BM25

```python
rankingBM25 = []
rank_temp = similaridadeBM25
for i in range(len(dados[0])):
  maior_rank = -10
  maior_index = 0
  for j in range(len(rank_temp)):
    if rank_temp[j] > maior_rank:
      maior_rank = rank_temp[j]
      maior_index = j
  rankingBM25.append(maior_index)
  rank_temp[maior_index] = -10
print(rankingBM25)
print("Lista dos documentos ordenados por ranking:")
for i in range(len(rankingBM25)):
  print(f"doc{rankingBM25[i]+1}.txt")
```
